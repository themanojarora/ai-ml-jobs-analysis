# AI Job Market Analysis - Geographic Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('ai_job_dataset.csv')
sns.set(style='whitegrid')
```

```python
# Top 10 countries by job postings
top_countries = df['company_location'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar')
plt.title('Top 10 Countries by Number of AI Job Postings')
plt.xlabel('Country')
plt.ylabel('Job Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Salary distribution across countries (Top 10 by frequency)
top_countries_list = top_countries.index.tolist()
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[df['company_location'].isin(top_countries_list)],
            x='company_location', y='salary_usd')
plt.title('Salary Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Remote job ratio by country
remote_by_country = df.groupby('company_location')['remote_ratio'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
remote_by_country.plot(kind='bar')
plt.title('Average Remote Ratio by Country (Top 10)')
plt.ylabel('Average Remote Ratio')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```
