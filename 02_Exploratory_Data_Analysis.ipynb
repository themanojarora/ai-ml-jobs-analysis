# AI Job Market Analysis - Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
df = pd.read_csv('ai_job_dataset.csv')
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_usd'], bins=50, kde=True)
plt.title('Distribution of Salaries (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='experience_level', y='salary_usd')
plt.title('Salary vs Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='remote_ratio')
plt.title('Remote Work Ratio Distribution')
plt.xlabel('Remote Ratio')
plt.ylabel('Job Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['job_title'].value_counts().plot(kind='bar')
plt.title('Job Title Frequency')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
skills_series = df['required_skills'].str.split(', ')
flat_skills = [skill for sublist in skills_series for skill in sublist]
top_skills = Counter(flat_skills).most_common(20)
skills_df = pd.DataFrame(top_skills, columns=['Skill', 'Count'])
plt.figure(figsize=(12, 6))
sns.barplot(data=skills_df, x='Count', y='Skill')
plt.title('Top 20 Required Skills')
plt.xlabel('Count')
plt.ylabel('Skill')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='company_size', y='salary_usd')
plt.title('Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
